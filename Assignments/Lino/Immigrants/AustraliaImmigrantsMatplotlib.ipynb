```python
# Plot bar chart of total no of immigrants for countries top 5 countires having max no of total immigrants
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel('AustraliaFiltered_LM.xlsx', index_col=0)
df.head()
```

```python
plt.subplots(figsize=(20,5))
df.Total.plot(kind='bar')
```

```python
df[['Country', 'Total']].sort_values(by='Total').tail()
```

```python
plt.subplots(figsize=(20,5))
plt.bar(df[['Country', 'Total']].sort_values(by='Total').tail().Country,
        df[['Country', 'Total']].sort_values(by='Total').tail().Total)
```

```python
# Find total immigrants from each category by area name and plot barchart for each area
df.groupby('AreaName')['Total'].sum()
```

```python
plt.subplots(figsize=(20,5))
plt.xlabel("AreaName", fontdict={'size':16})
plt.ylabel("Total immigrants in Millions", fontdict={'size':16})
plt.xticks(fontsize=12)
df.groupby('AreaName')['Total'].sum().plot(kind='bar',rot=0)
```

```python
# Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).
df[df.AreaName=='Asia'].sum(numeric_only=True)[:-1]  # omitting last column 'Total', hence [:-1] is applied 
```

```python
plt.subplots(figsize=(20,5))
plt.xlabel("year", fontdict={'size':16})
plt.ylabel("Total count", fontdict={'size':16})

plt.plot(df.columns[4:33],df[df.AreaName=='Asia'].sum(numeric_only=True)[:-1],marker='o', label ='Asia')
plt.plot(df.columns[4:33],df[df.AreaName=='Africa'].sum(numeric_only=True)[:-1],marker='o', label ='Africa')
plt.plot(df.columns[4:33],df[df.AreaName=='Europe'].sum(numeric_only=True)[:-1],marker='o', label ='Europe')
plt.legend()
```

```python
# Plot Total Immigrants based on DevelopmenRegion
df.groupby('DevName')['Total'].sum()
```

```python
# just confirming the Total count, since both counts are too close
df[df['DevName']=='Developed regions']['Total'].sum(), df[df['DevName']=='Developing regions']['Total'].sum()
```

```python
plt.ylabel("Total count in millions", fontdict={'size':16})
df.groupby('DevName')['Total'].sum().plot(kind='bar', rot=0)
```

```python
# Plot immigration trend for India using line plot
plt.subplots(figsize=(20,5))
plt.xlabel("year", fontdict={'size':16})
plt.ylabel("Total count", fontdict={'size':16})

plt.plot(df.columns[4:33],df[df.Country=='India'].iloc[0,4:33], marker='o')
```
