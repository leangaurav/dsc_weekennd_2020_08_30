```python
import functools
def img_mapper(x):
    ret = True
    if(x == 0):
        ret = False
    return ret

def xor_c(row):
    list_row = list(row)
    return not functools.reduce(lambda a,b:a and b, list_row)

df[df['Development region'].iloc[:,1:].applymap(img_mapper).apply(xor_c, axis=1)]['Origin/Destination']['Country'].unique()
```
