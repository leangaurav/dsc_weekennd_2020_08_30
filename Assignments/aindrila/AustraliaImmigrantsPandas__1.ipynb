```python
#List all Column and Row indexes(Print their names)
an=pd.read_excel("Australia.xlsx",sheet_name='Australia by Residence',skiprows=20)
print(an)
print(an.columns)
for row in an.index: 
    print(row, end= " ") 
```

```python
#Use values and index attributes on both Row and Column Indexes
an.columns.values[0] = 'Type'
an = an.rename(index = lambda x: x + 10) 
  
print(an)
```

```python
#Check Shape, Datatypes and other attributes and observe the datatype of each columns' data

result = an.dtypes 
  
# Print the result 
print(result)
print("Info" ,an.info)
print()
print("Length:" ,len(an))
print()
print("shape: ",an.shape)
print()
print("Size:," ,an.size)
```

```python
row, col = an.shape
print(row)
# 891

print(col)
# 12
```

```python
#Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)


index_names = an[ (an['Type'] == 'Emigrants') | (an['Type'] != 'Immigrants')].index 
  
# drop these given row 
# indexes from dataFrame 
an.drop(index_names, inplace = True)

print(an)
```

```python
#Find columns for which more than 50% data is missing and drop them (Column Deletion)
missing = an.isnull().sum()
print(missing)
        
#print(an.shape)
```

```python
column_with_nan = an.columns[an.isnull().any()]
print(an)
```

```python
#Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion)
#    Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"

ag=an.drop(['Type','Coverage','AREdf.drop('column_name', axis=1, inplace=True)','REG','DEV'], axis = 1) 
print(ag)
```

```python
#- Rename columns :OdName -> Country; RegName -> Region
an.rename(columns = {'OdName':'Country', 'RegName':'Region'}, inplace = True) 
   
# After renaming the columns 
print(an.columns) 
```

```python
#- Convert Column Names with Int type to 
#Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)

an['1980'] = an['1980'].astype('str')
an['1980'].dtype
result = an.dtypes 
  
# Print the result 
print(result)
```

```python
#Find Total number of immigrants for all years, and create a new column Total in data Frame

subset_df = an[an["Type"] == 'Immigrants']
#column_sum = subset_df['1981'].sum(axis=0)
fg=subset_df.iloc[:, 8:43]
an['Total']=fg.sum(axis=1)
print(an)

#subset_df['Fruit Total']=subset_df.iloc[:,-4:].sum(axis=1)
```

```python
an.drop('Fruit Total', axis=1, inplace=True)

print(an)
```

```python
an.sort_values(by=['Total'], inplace=True, ascending=False)
print(an)
```

```python
world_df=an[an.AreaName == 'World']
indexNames = an[an['AreaName'] == 'World' ].index
an.drop(indexNames , inplace=True)
#print(an)
world_df=an[an.AreaName == 'World']
print(an)
# Delete these row indexes from dataFrame
#dfObj.drop(indexNames , inplace=True)
an.to_excel('AustraliaFiltered.xlsx', index = False)
```
