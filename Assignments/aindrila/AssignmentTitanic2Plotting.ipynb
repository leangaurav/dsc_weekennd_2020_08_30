
### Titanic Plotting
- Load data generated previously from 'titanic_filtered.csv'

- plot a horizontal bar-chart depicting the ratio of males to females
- plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead. 
- plot a stacked bar chart of number of male-survived,dead and female-survived, dead.

- plot box plot of ages of survived vs not-survived

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# plot histogram of ages of people
import pandas as pd
import numpy as np
tf=pd.read_csv("titanic_dataset.csv",delimiter = ',')
tf.age.plot(kind="hist",width=1)
```

```python
#plot no of people in each pclass on bar chart

r=tf['pclass'].value_counts()
r.plot(kind="bar")
```

```python
# plot a bar-chart depicting the ratio of males to females
tf['gender'].value_counts()
mf=tf['gender'].value_counts()
#mf.plot(kind="bar")
mf.plot.barh()
```

```python
# plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.
fg=tf.groupby(['gender', 'survived']).size()
fg.plot.bar()
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
fg=tf.groupby(['gender', 'survived']).size().unstack()
fg.plot(kind='bar', stacked=True, figsize=(10, 5))
```

```python
# plot stacked bar of survived with different pclass in same bar

 
fg=tf.groupby(['pclass', 'survived']).size().unstack()
fg1=tf.groupby(['survived', 'pclass']).size().unstack()
fg.plot(kind='bar', stacked=True, figsize=(10, 5))
fg1.plot(kind='bar', stacked=True, figsize=(10, 5))
```

```python
# plot box plot of ages of survived vs not-survived
fg2=tf.groupby(['survived', 'age']).size().unstack()
fg2.head()
tf.boxplot(by ='survived', column =['age'], grid = False) 
 
```

```python
# plot scatter plot of ages vs survived.
plt.scatter(tf.age,tf.survived,alpha=0.4)
```
