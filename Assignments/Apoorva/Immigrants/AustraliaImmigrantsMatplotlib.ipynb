```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel('AustraliaFiltered_2.xlsx')
```

```python
df1
```

#### Plot bar chart of total no of immigrants for countires top 5 countires having max no of total immigrants

```python
df2 = df1.sort_values(by="Total",ascending = False).head(5).loc[:,["Total","Country"]]
plt.bar(df2.Country, df2.Total)
plt.xticks(rotation = 80)
plt.show()
```

#### Find total immigrants from each category by area name and plot barchart for each area

```python
grp = df1.groupby(by="AreaName")
grp.count() # Data Frame
print(grp.count().Country)
plt.bar(grp.count().Country.index, grp.count().Country.values)
plt.xticks(rotation=45)
plt.show()
```

#### Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot.

```python
df2 = df1.iloc[:,5:34]
s2 = df2.sum()
plt.plot(s2.index,s2.values)
plt.xticks(rotation=90)
plt.show()
```

#### Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).

```python
s = df1[df1["AreaName"]=="Asia"].Total
s1 =  df1[df1["AreaName"]=="Africa"].Total
s2 = df1[df1["AreaName"]=="Europe"].Total
plt.plot(s.index, s.values, s1.index, s1.values, s2.index, s2.values)
plt.show()
```

#### Plot immigration trend for India using line plot

```python
s = df1[df1["Country"]=="India"].iloc[0,5:34]
plt.plot(s.index,s.values)
plt.xticks(rotation=90)
plt.show()
```

#### Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot

```python
df = df1.sort_values(by="Total",ascending = False).head(5).loc[:,["Total","Country"]]
df
plt.plot(df.Country,df.Total)
plt.xticks(rotation=90)
plt.show()
```

#### For the above use area plot with alpha = 0.5


```python
df = df1.sort_values(by="Total",ascending = False).head(5).loc[:,["Total","Country"]]
df.plot.area(x="Country",y="Total",alpha=0.5)
#plt.plot(df.Country,df.Total)
plt.xticks(rotation=90)
plt.show()
```

#### Compare Immigration Trend for top-5 countires using stacked area plot

```python
df = df1.sort_values(by="Total",ascending = False).head(5).loc[:,["Total","Country"]]
#df.plot.area(x="Country",y="Total",alpha=0.5)
df.plot.area(x="Country",y="Total",stacked=True)
plt.show()
```
