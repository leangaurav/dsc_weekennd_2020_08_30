####    Read sheet "Australia by Residence" from Australia.xlsx
####    Ignore the first 20 rows.
####    List all Column and Row indexes(Print their names)

```python
import pandas as pd
import numpy as np
df1 = pd.read_excel('Australia.xlsx' ,sheet_name='Australia by Residence', skiprows= [i for i in range(0,20)])
print(df1.shape)
print('Row Index =>',df1.index)
for i in df1.index:
    print(i, end=", ")
print()
print('Column Index =>',df1.columns)
for i in df1.columns:
    print(i, end=", ")
```

```python
print("shape: ",df1.shape , " size: ",df1.size)
print(df1.dtypes)
print()
```

### Drop

#### Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)

```python
df2 = df1
print(df1[df1.Type == 'Immigrants'])
df1 = df1[df1.Type == 'Immigrants']
```

#### Find columns for which more than 50% data is missing and drop them (Column Deletion)

```python
df2 = df1.dropna( axis=1 , thresh=0.50*df1.shape[0] , inplace=True )
print(df1)
```

#### Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"

```python
df1.drop(['Type','Coverage','AREA','REG','DEV'],axis=1,inplace=True)
df1
```

### Rename

#### Rename columns :OdName -> Country; RegName -> Region

```python
df1.rename({'OdName':'Country', 'RegName':'Region'},axis = 1,inplace=True)
df1
```

#### Convert Column Names with Int type to Strings to avoid index errors with string indexing "1980, 1981 etc." => " '1980', '1981' ... " (Use rename)

```python
print(df1.columns)
df1.rename(lambda x: str(x) ,axis = 1,inplace=True)
print(df1.columns)
```
