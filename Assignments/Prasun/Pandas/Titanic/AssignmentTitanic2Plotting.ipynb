
### Titanic Plotting
- Load data generated previously from 'titanic_filtered.csv'
- plot histogram of ages of people
- plot no of people in each pclass on bar chart


- plot a horizontal bar-chart depicting the ratio of males to females
- plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead. 
- plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
- plot stacked bar of survived with different pclass in same bar.


- plot box plot of ages of survived vs not-survived
- plot scatter plot of ages vs survived.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# plot histogram of ages of people
```

```python
tt=pd.read_csv('titanic_filtered.csv')
tt.age.plot(kind='hist')
```

```python
#plot no of people in each pclass on bar chart
tt.groupby('pclass').size().plot(kind='bar')
```

```python
tt.groupby('gender').size().plot(kind='barh')
```

```python
# plot a bar-chart depicting the ratio of males to females
```

```python
tt.groupby(['survived','gender']).size().plot(kind='bar')
```

```python
# plot a bar-chart depicting the ratio of male-survived, female-survived, male-dead, female-dead.
```

```python
tt.groupby(['gender','survived']).size().unstack('survived').plot(kind='bar',stacked=True)
```

```python
# plot a stacked bar chart of number of male-survived,dead and female-survived, dead.
```

```python
# plot stacked bar of survived with different pclass in same bar


tt.groupby(['pclass','survived'])['pclass'].size().unstack('survived').plot(kind='bar',stacked=True)
```

```python
tt.groupby(['pclass','survived'])['survived'].size().unstack('pclass').plot(kind='bar',stacked=True)
```

```python
# plot box plot of ages of survived vs not-survived

tt.boxplot(by='survived',column='age')
```

```python
tt.boxplot(by='survived',column='age')
```

```python
# plot scatter plot of ages vs survived.

plt.scatter(tt.age,tt.survived)
plt.show()
```

```python
sur_grp= tt.groupby(by =['gender','survived'])
sur_grp.size().plot(kind='bar',color=['yellow','blue','green','red'])
```
