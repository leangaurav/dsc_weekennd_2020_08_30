- Find total immigrants from each category by area name and plot barchart for each area

- Compare Immigration Trend for top-5 countires using stacked area plot

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
aus=pd.read_excel('AustraliaFiltered.xlsx')
aus.head()
```

```python
aus.sort_values(by=['Total'],ascending=False)['Total'].head().plot(kind='bar')
```

```python
aus.groupby('AreaName')['Total'].size().plot(kind='bar')
```

```python
#Plot total immigration trend(immigration of world over the years i.e. year vs Total count for that year) using line plot.
print(aus[[str(i) for i in range(1980,2009)]].sum())
aus[[str(i) for i in range(1980,2009)]].sum().plot()
```

```python
#Compare Immigration Trend for Asia, Africa and Europe AreaName (Use Line plot).
aus4 = aus.groupby(by='AreaName')[[str(i) for i in range(1980,2009)]].sum().T.loc[:,'Africa':'Europe']
aus4.plot()
```

```python
#Plot Total Immigrants based on DevelopmenRegion
#aus.Total.plot()
aus.groupby(by=['DevName'])['Total'].sum().plot(kind='bar')
```

```python
#Plot immigration trend for India using line plot
aus1=aus.loc[aus['Country']=='India'][[str(i) for i in range(1980,2009)]].T
aus1.plot()
```

```python
#Plot immigration Trend(year vs count) for the coutries with top 5 total immigrants. Use Line plot
aus.sort_values(by='Total', ascending=False).head().groupby(by='Country')[[str(i) for i in range(1980,2009)]].sum().T.plot()
```

```python
#For the above use area plot with alpha = 0.5
aus.sort_values(by='Total', ascending=False).head().groupby(by='Country')[[str(i) for i in range(1980,2009)]].sum().T.plot(kind='area', alpha=0.5, stacked=False)
```

```python
#Compare Immigration Trend for top-5 countires using stacked area plot
aus.sort_values(by='Total', ascending=False).head().groupby(by='Country')[[str(i) for i in range(1980,2009)]].sum().T.plot(kind='area',alpha=0.5,stacked=True)
```
