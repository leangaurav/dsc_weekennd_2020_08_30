```python
#print(read_excel.isin([0,'..']).mean())
read_excel = read_excel.loc[:,read_excel.isin([0,'..']).mean()<0.5]
read_excel.head()
```

```python
read_excel= read_excel.drop(labels=['Type','Coverage','AREA','REG','DEV'],axis=1)
read_excel.head()
```

```python
read_excel= read_excel.rename(columns={'OdName':'Country','RegName':'Region'})
read_excel.head()
```

```python
read_excel = read_excel.rename(str, axis=1)
#print(read_excel['1980'])
#read_excel.head()
Total = read_excel.sum(axis=1, numeric_only=True)
read_excel["Total"]= Total
read_excel.head()
```

```python
read_excel1 = read_excel.sort_values(by=["Total"],axis=0, ascending=False)
read_excel1 = read_excel1.drop(labels=[448], axis=0)
read_excel.tail()
read_excel = read_excel.drop(labels=[447, 448], axis=0)
read_excel.tail()
```
