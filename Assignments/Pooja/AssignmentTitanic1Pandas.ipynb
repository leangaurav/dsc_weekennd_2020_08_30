```python
import pandas as pd
titanic=pd.read_csv('titanic_dataset.csv')
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.isna().any()
```

```python
titanic.shape
```

```python
import numpy as np
#print(titanic.isin([np.nan,' ']).mean())
titanic = titanic.loc[:, titanic.isin([' ',np.nan]).mean()<.25]
titanic.head()
```

```python
titanic= titanic.drop(columns=['name','ticket'],axis=1)
titanic.head()
```

```python
titanic.dtypes
```

```python
import re
titanic.fare = pd.to_numeric(titanic.fare.str.replace(r'\$',''))
```

```python
titanic.dtypes
```

```python
titanic.isna().sum()
```

```python
titanic.fillna({'age': titanic.age.mean() , 'fare':titanic.fare.mean()})
```

```python
titanic.dropna(subset= ['embarked'], inplace = True)
titanic.head()
```

```python
titanic.to_csv('Titanic_Filtered_PR.csv')
```

```python
titanic['survived'].replace({0:'D',1:'A'},inplace=True)
titanic.head()
```
