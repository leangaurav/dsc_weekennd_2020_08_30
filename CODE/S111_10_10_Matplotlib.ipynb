```python
import json
class Point:
    def __init__(self, x,y):
        self.x = x
        self.y = y
        
    def __str__(self):
        return f"x={self.x} y={self.y}"

    def to_json(self):
        return json.dumps(self.__dict__)
    
    @staticmethod
    def from_json(string):
        js = json.loads(string)
        return Point(js["x"], js["y"])

p1= Point(10,20)
```

```python
print(p1)
s = p1.to_json()
print(s)
```

```python
p2 = Point.from_json(s)
print(p2)
print(id(p1), id(p2))
```
